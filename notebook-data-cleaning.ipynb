```python
folder_path = "assets/data"

filenames = [f for f in os.listdir(folder_path)]

def parse_date(filename):
    name = filename.replace(".csv", "")
    day, month = name.split()
    for fmt in ("%d %b %Y", "%d %B %Y"):
        try:
            return datetime.strptime(f"{day} {month} 2025", fmt)
        except ValueError:
            continue
    raise ValueError(f"Cannot parse date from {filename}")

sorted_files = sorted(filenames, key=parse_date)

for idx, f in enumerate(sorted_files, 1):
    date_obj = parse_date(f)
    new_name = f"{date_obj.day:02d} {date_obj.strftime('%B')}.csv"  # full month name
    old_path = os.path.join(folder_path, f)
    new_path = os.path.join(folder_path, new_name)
    os.rename(old_path, new_path)
    print(f"Renamed '{f}' → '{new_name}'")
```

```python
col_map = {
    "Product": ["Product Name", " ", "Products", "3", "1"],
    "Unit": ["Unit", "Units"],
    "Purchase Price per Unit": ["Purchase Price Per Unit (Actual)", "PP Per Unit (Actual)", "Actual PP"],
    "Selling Price per Unit": ["Our Selling Price Per Unit", "Seling Price", "SP"],
    "Kalimati Min": ["Kalimati Rate Minimum", "Kalimati Minimum", "Kalimati Mini"],
    "Kalimati Max": ["Kalimati Rate Maximum", "Kalimati Maximum", "Kalimati Maxi"],
    "Kalimati Average Price": ["Kalimati Average Price", "KAVG Rate"],
    "Unit Sales": ["Units Sales"],
    "Unit Purchase": ["Purchase", "Purchase Units", "Purchs Units"],
    "Total Cost": ["Total cost"],
    "Total Sales": ["Total Sales"]
}
    # "Closing Stock": ["Closing Stock", "CS"],
    # "Closing Stock Value": ["CS Value", "Closing Stock Value"]
```

### Merging the datasets

```python
all_data = []
for filename in filenames_sorted:
    df = pd.read_csv(os.path.join(folder_path, filename))
    
    df.columns = df.columns.str.strip()
    
    rename_dict = {}
    for key, aliases in col_map.items():
        for alias in aliases:
            alias_stripped = alias.strip()
            if alias_stripped in df.columns:
                rename_dict[alias_stripped] = key
                break
    
    df.rename(columns=rename_dict, inplace=True)
    

    available_cols = [c for c in df.columns if c in col_map.keys()]
    df = df[available_cols]
    
    name = filename.replace(".csv", "")
    date = datetime.strptime(name + " 2025", "%d %B %Y")
    df["Date"] = date.strftime("%Y-%m-%d")
    
    all_data.append(df)

data = pd.concat(all_data, ignore_index=True)
```

### Filling Missing Values

```python
data["Total Sales"] = pd.to_numeric(data["Total Sales"].astype(str).str.replace(",", ""), errors='coerce')
data["Selling Price per Unit"] = pd.to_numeric(data["Selling Price per Unit"].astype(str).str.replace(",", ""), errors='coerce')

data["Total Cost"] = pd.to_numeric(data["Total Sales"].astype(str).str.replace(",", ""), errors='coerce')
data["Purchase Price per Unit"] = pd.to_numeric(data["Selling Price per Unit"].astype(str).str.replace(",", ""), errors='coerce')
```

```python
data["Unit Sales"] = data["Unit Sales"].fillna(data["Total Sales"] // data["Selling Price per Unit"])
```

```python
data["Unit Purchase"] = data["Unit Purchase"].fillna(data["Total Cost"] // data["Purchase Price per Unit"])
```
